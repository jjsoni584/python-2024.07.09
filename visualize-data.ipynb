### 그래프의 종류
**막대그래프** : 데이터를 카테고리별로 값을 막대로 표현 할때 사용하는 그래프 형식(수직 막대 그래프, 수평 막대 그래프)
> 카테고리별 데이터를 비교하고자 할때 사용
>(제품 판매량, 인구수, 지역별)
** 선 그래프 **: 시간의 흐름에 따른 데이터 포인터를 선으로 연결하여 사용하는 그래프 형식
> 하나의 요소에 대해 시간의 흐름에 따라 변경값을 나타내고자 할때 사용(기온, 주식)
**파이 차트** :전체 데이터에 대한 각 요소의 비율을 원형으로 나타낼때 사용하는 그래프 형식
> 전체에 대한 요소의 비율을 표현하고자 할때 사용(튜표결과)
**산점도** : 두 변수 간의 상관관계를 표현할때 사용하는 그래프 형식
>각 요소간 관계성을 파악할때 사용(키와 몸무게 상관관계 비교)
**히스토그램** : 범위별로 그룹화하여 각 그룹에 속하는 데이터의 수를 막대로 표현하는 그래프 형식
> 데이터의 분포를 비교할때 사용(시험 점수 분포)
**영역 그래프** : 선 그래프의 아래쪽 빈 공간을 채운 그래프
>시간의 흐름에 따라 데이터를 비교할 때 사용, 변화량을 강조하고자 할때 사용
''박스 그래스,' : 데이터의 분포를 나타낼때 사용하는 그래프
>최댓값, 최소값 사분위주로 데이터의 이상치를 판단할때 사용

### 선 그래프
가장 기본적인 그래프 형식, 시간의 흐름 및 순서에 따라서 데이터의 변화량을 보여주고자 할 때 사용
plot패키지의 plot에서 메서드로 생성 가능

```python
plt.plot([1,2,4,8,16])
plt.show()
```

```python
# 그래프의 제목 지정 title() 메서드 사용
#plot(X축에 대한 리스트, y축에 대한 리스트)
plt.title('x축 추가')
plt.plot([2020, 2021, 2022, 2023, 2024], [1,2,4,8,16])
plt.show
```

### 막대 그래프
x축의 값이 분류가 가능한 카테고리 형태일때 bar(), barh() 메서드로 막대 그래프를 표현할 수 있음

```python
plt.title('막대 그래프')
plt.bar(['a','b','c'],[20, 78, 43], color='#ff0000')
plt.show
```

```python
# 가로 막대그래프
# barh() 메서드로 가로 막대그래프를 표현 가능
# 오차 표시
# xerr, yerr 매개변수로 오차를 지정할 수 있음
plt.title('막대 그래프')
plt.barh(['a','b','c'],[20, 78, 43], xerr=[2, 10, 4],color='#ff0050')
plt.show
```

### 스템 플롯
기준점을 기준으로 오차범위를 표현하고자 할때 사용하는 차트

```python
x = np.linspace(0.1, 2 * np.pi, 10)
plt.title('스템 플롯')
plt.stem(x, np.cos(x))
plt.show()
```

### 파이 차트
각 분류가 전체에서 차지하는 비중을 표현하고자 할때 사용하는 그래프

```python
plt.title('파이차트')
plt.pie([25, 45, 30], labels=['사과', '바나나', '포토'],
    explode= [0, 0.1, 0], colors=['#ff0050', '#00ff0050', '#000ff550'],
    shadow=True)
plt.axis('equal')
plt.show()
```

### 컨투어 플롯
3차원 자료에서 x,y 입력과 z출력이 존재할 때, z를 선과 색상으로 표현하는 그래프

```python
def func(x, y):
    z = (1-x / 2+x ** 5 + y ** 3) * np.exp(-x ** 2 -y ** 2)
    return z

x = np.linspace(-3, 3, 256)
y = np.linspace(-3, 3, 256)
xx, yy = np.meshgrid(x, y)
zz= func(xx, yy)

plt.title('등고선 그래프')
plt.contour(xx, yy, zz, colors='black') 
plt.contourf(xx, yy, zz)
plt.grid(True)
plt.show()
```
